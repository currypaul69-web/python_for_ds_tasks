```python
def show_menu (страви):
    print("=== Меню ===")
    for номер, страва in enumerate(страви, start=1):
        print(f"{номер}. {страва}")
    print("=============")
show_menu(["Суп грибний", "Котлета по-київськи", "Лимонад"])
```

```python
def print_square (n):
    for  квадрат in range(1, n+1):
        print(f"Квадрат числа {квадрат} = {квадрат**2}")
print_square(5)
```

```python

def multiply (a, b):
    return a * b
print(multiply(10, 3))
```

```python
def find_max (numbers):
    max_num=numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num
    return max_num
print(find_max([15, 42, 8, 23, 56, 19]))
```
