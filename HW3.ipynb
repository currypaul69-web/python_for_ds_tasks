```python
cooking_time = 5
if cooking_time < 4:
    print ("Яйце рідке")
elif 4 <= cooking_time < 7:
    print ("Яйце всмятку")
elif 7 <= cooking_time < 10:
    print ("Яйце вкруту")
else:
    print ("Яйце переварене")
```

```python
grade = 72
if grade >= 90:
    print ("Відмінно")
elif 75 <= grade < 90:
    print ("Добре")
elif 60 <= grade < 75:
    print ("Задовільно")
else:
    print ("Незадовільно")
```
